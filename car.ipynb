```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
features =["high-mpg","city-mpg","symboling","compression-ratio","height","bore", "wheel_base", "length", "horsepower",  "engine-size", "width", "curb-weight"]       
X = df[features]
Y = df[["price"]]

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=.5,random_state=1)

print("number of test sample: ",x_test.shape[0])
print("number of training samples: ",x_train.shape[0])
```

```python
from sklearn.linear_model import Ridge
```

```python
RM = Ridge(alpha=0.1)
RM.fit(x_train,y_train)
RM.score(x_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pr = PolynomialFeatures(degree=2,include_bias=False)
polly = pr.fit_transform(x_train,y_train)
polly2 = pr.fit_transform(x_test,y_test)
```

```python
RM2 = Ridge(alpha=0.1)
RM2.fit(polly,y_train)
RM2.score(polly2,y_test)
```
